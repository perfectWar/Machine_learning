```python
import requests 
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = requests.get('https://www.imdb.com/title/tt27995594/reviews?ref_=tt_urv')
```

```python
req
```

```python
req.content
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
print(soup.prettify())
```

# Scrap movie name

```python
scrapped_movies = soup.find_all('div',class_='parent')
```

```python
scrapped_movies
```

```python
movies = []
for movie in scrapped_movies:
    movie = movie.get_text().replace('\n','')
    movie = movie.strip(" ")
    movies.append(movie)
```

```python
movies
```

# scrap review

```python
review_columns = soup.find_all("div", class_="text show-more__control")
review_ratings = soup.find_all("span", class_="rating-other-user-rating")

review_list = []
for column in review_columns:
    review_list = column.get_text().strip()

ratings_list = [rating.get_text(strip=True) for rating in review_ratings]
```

```python
review_list
ratings_list
```

```python
len(review_list)
len(ratings_list)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"Review": review_list, "Rating": ratings_list})
```

```python
df.to_csv("imdb_reviews.csv", index=False)
```

```python
df = pd.read_csv('imdb_reviews.csv')
```

```python
df.head()
```

```python
def extract_rating(rating):
    return rating.split('/')[0]
```

```python
df['Rating'] = df['Rating'].apply(extract_rating)
```

```python
df.head()
df['Rating'][0]
```

```python
def analysis(rating):
    if int(rating) > 5:
        return 'Positive'
    elif int(rating) < 5:
        return 'Negative'
    return 'Neutral'
```

```python
df['Sentiment'] = df['Rating'].apply(analysis)
```
