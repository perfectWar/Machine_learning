# Data reading

```python
#importing the library
import pandas as pd
import numpy as np
```

```python
#reading the csv file
df = pd.read_csv("C:\\Users\\raths\\Downloads\\Attrition - Attrition.csv")
#top 1 data of csv file 
df.head(1)
```

```python
#removing row which have null valuesn
df = df.dropna()
#columns names 
columns = df.columns
columns
```

# LabelEncoding

```python
#import the label encoder library
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
#string type of data columns 
df[['BusinessTravel','Attrition','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']]
```

```python
#applying the labelEncoding to the string datatype columns 
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['Attrition'] = le.fit_transform(df['Attrition'])
df['Department'] = le.fit_transform(df['Department'])
df['EducationField'] = le.fit_transform(df['EducationField'])
df['Gender'] = le.fit_transform(df['Gender'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['Over18'] = le.fit_transform(df['Over18'])
df['OverTime'] = le.fit_transform(df['OverTime'])
```

```python
#checking the output of any 3 data of dataset
df.sample(3)
```

# Train_test_split

```python
x = df.drop(columns = ['Attrition'], axis= 1) #independent data
y = df['Attrition'] #dependent data 
```

```python
x.shape #shape of independent data 
```

```python
y.shape #shape of dependent data
```

```python
#importing train test split library 
from sklearn.model_selection import train_test_split
```

```python
#80% of data to training and 20% of data to testing 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=9) 
```

```python
x_train.shape #shape of training data of independent data
```

```python
x_test.shape #shape of testing data of independent data 
```

```python
y_train.shape #shape of training data of dependent data 
```

# Standardization ===> Data Mean = 0, Standard Deviation = 1

```python
#importing the Standardization library
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train) #fit is to learn the parameter and transform is to aplly the changes 
# x_train_sc
```

```python
x_train_new = pd.DataFrame(x_train_sc, columns = x_train.columns) #array to the dataframe
# x_train_new
```

```python
np.round(x_train_new.describe(),1) #checking the output of the standardization by 1 place of decimal 
```
