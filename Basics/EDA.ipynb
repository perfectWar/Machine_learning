# Parts of EDA

1. univariate Analysis ===> analysis on a single independent column
2. Bivariate Analysis ===> analysis on two columns
3. multivariate analysis ===> on more than 2 columns

data type 
1. numerical ===> continuous ===> age(year,month, days)
2. categorical dat ===> descrete ===> total number of employee
